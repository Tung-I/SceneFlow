```python
def bilinear_interp_val(vmap, y, x):
    '''
        bilinear interpolation on a 2D map
    '''
    h, w = vmap.shape
    x1 = int(x)
    x2 = x1 + 1
    x2 = w-1 if x2 > (w-1) else x2
    y1 = int(y)
    y2 = y1 + 1
    y2 = h-1 if y2 > (h-1) else y2
    Q11 = vmap[y1,x1]
    Q21 = vmap[y1,x2]
    Q12 = vmap[y2,x1]
    Q22 = vmap[y2,x2]
    return Q11 * (x2-x) * (y2-y) + Q21 * (x-x1) * (y2-y) + Q12 * (x2-x) * (y-y1) + Q22 * (x-x1) * (y-y1)


def get_3d_pos_xy(y_prime, x_prime, depth, focal_length=1050., w=960, h=540):
    '''
        depth pop up
    '''
    y = (y_prime - h / 2.) * depth / focal_length
    x = (x_prime - w / 2.) * depth / focal_length
    return [x, y, depth]
```

```python
def gen_data(fname_disparity, fname_disparity_next_frame, fname_disparity_change, fname_optical_flow, \
                  image_l, image_next_frame_l, image_r, image_next_frame_r, n = 8192, max_cut = 35, focal_length=1050.):

    np.random.seed(601985)

    ##### generate needed data
    disparity_np, _ = load_pfm(fname_disparity)
    disparity_next_frame_np, _ = load_pfm(fname_disparity_next_frame)
    disparity_change_np, _ = load_pfm(fname_disparity_change)
    optical_flow_np, _ = load_pfm(fname_optical_flow)
    rgb_l = imageio.imread(image_l)[:, :, :-1] / 255.
    rgb_next_frame_l = imageio.imread(image_next_frame_l)[:, :, :-1] / 255.
    rgb_r = imageio.imread(image_r)[:, :, :-1] / 255.
    rgb_next_frame_r = imageio.imread(image_next_frame_r)[:, :, :-1] / 255.

    depth_np = focal_length / disparity_np
    depth_next_frame_np = focal_length / disparity_next_frame_np
    future_depth_np = focal_length / (disparity_np + disparity_change_np)
    
    h, w = disparity_np.shape

    ##### flow generation
    
    optical_flow_x = np.array(optical_flow_np[:, :, 0])
    optical_flow_y = np.array(optical_flow_np[:, :, 1])
    depth_flow = future_depth_np - depth_np
    flow = np.stack((optical_flow_x, optical_flow_y, depth_flow), 2)

    ##### only take pixels that are near enough
    try:
        depth_requirement = depth_next_frame_np < max_cut
    except:
        return None

    satisfy_pix2 = np.column_stack(np.where(depth_next_frame_np < max_cut))
    if satisfy_pix2.shape[0] < n:
        return None
    sample_choice2 = np.random.choice(satisfy_pix2.shape[0], size=n, replace=False)
    sampled_pix2_y = satisfy_pix2[sample_choice2, 0]
    sampled_pix2_x = satisfy_pix2[sample_choice2, 1]
    

    ##### flip
    rgb_l = np.flip(rgb_l, 0)
    rgb_r = np.flip(rgb_r, 0)
    rgb_next_frame_l = np.flip(rgb_next_frame_l, 0)
    rgb_next_frame_r = np.flip(rgb_next_frame_r, 0)
        
    
    ##### normalize
#     disparity_np = (disparity_np - np.mean(disparity_np)) / np.var(disparity_np)
#     disparity_next_frame_np = (disparity_next_frame_np - np.mean(disparity_next_frame_np)) / np.var(disparity_next_frame_np)
#     flow = (flow - np.mean(flow)) / np.var(flow)
    
    ##### check NaN value
    rgb_l[np.where(np.isnan(rgb_l)!=0)] = 0
    rgb_r[np.where(np.isnan(rgb_r)!=0)] = 0
    rgb_next_frame_l[np.where(np.isnan(rgb_next_frame_l)!=0)] = 0
    rgb_next_frame_r[np.where(np.isnan(rgb_next_frame_r)!=0)] = 0
    disparity_np[np.where(np.isnan(disparity_np)!=0)] = 0
    disparity_next_frame_np[np.where(np.isnan(disparity_next_frame_np)!=0)] = 0
    flow[np.where(np.isnan(flow)!=0)] = 0
#     mask[np.where(np.isnan(mask)!=0)] = 0

        
    return rgb_l, rgb_r, rgb_next_frame_l, rgb_next_frame_r, disparity_np, disparity_next_frame_np, flow
```

```python
def proc_one_scene(s, input_dir_1, input_dir_2, output_dir):
    if s[-1] == '/':
        s = s[:-1]
    dis_split = s.split('/')
    train_or_test = dis_split[-4]
    ABC = dis_split[-3]
    scene_idx = dis_split[-2]
    left_right = dis_split[-1]
    for v in range(6, 15):
        fname = os.path.join(output_dir, train_or_test + '_' + ABC + '_' + scene_idx + '_' + left_right + '_' + str(v).zfill(4) + '-{}'.format(0) + '.npz')

        fname_disparity = os.path.join(input_dir_1, 'disparity', train_or_test, ABC, scene_idx, left_right, str(v).zfill(4) + '.pfm')
        fname_disparity_next_frame = os.path.join(input_dir_1, 'disparity', train_or_test, ABC, scene_idx, left_right, str(v+1).zfill(4) + '.pfm')
        fname_disparity_change = os.path.join(input_dir_2, 'disparity_change', train_or_test, ABC, scene_idx, 'into_future', left_right, str(v).zfill(4) + '.pfm')
        
        fname_image_l = os.path.join(input_dir_1, 'frames_finalpass', train_or_test, ABC, scene_idx, left_right, str(v).zfill(4) + '.png')
        fname_image_next_frame_l = os.path.join(input_dir_1, 'frames_finalpass', train_or_test, ABC, scene_idx, left_right, str(v+1).zfill(4) + '.png')
        
        fname_image_r = os.path.join(input_dir_1, 'frames_finalpass', train_or_test, ABC, scene_idx, 'right', str(v).zfill(4) + '.png')
        fname_image_next_frame_r = os.path.join(input_dir_1, 'frames_finalpass', train_or_test, ABC, scene_idx, 'right', str(v+1).zfill(4) + '.png')
        
        L_R = 'L'
        fname_optical_flow = os.path.join(input_dir_2, 'optical_flow', train_or_test, ABC, scene_idx, 'into_future', left_right, 'OpticalFlowIntoFuture_' + str(v).zfill(4) + '_' + L_R + '.pfm')

        d = gen_data(fname_disparity, fname_disparity_next_frame, fname_disparity_change, fname_optical_flow, \
                          fname_image_l, fname_image_next_frame_l, fname_image_r, fname_image_next_frame_r, focal_length=1050.)
        
        if d is not None:
            np.savez_compressed(fname, rgb_l=d[0], \
                                       rgb_r=d[1], \
                                       rgb_next_l=d[2], \
                                       rgb_next_r=d[3], \
                                       disparity=d[4], \
                                       disparity_next=d[5], \
                                       flow=d[6])
        else:
            print(fname)
```

```python
INPUT_DIR_1 = '/datasets'
INPUT_DIR_2 = '/data'
OUTPUT_DIR = '/home/tony/datasets/flyingthings3d_preprocessed'

if not os.path.exists(OUTPUT_DIR):
    os.system('mkdir -p {}'.format(OUTPUT_DIR))

random.seed('ECCV')

disparity_scenes_train = sorted(list(glob.glob(os.path.join(INPUT_DIR_1, 'disparity/TRAIN/*/*/*/'))))
disparity_scenes_test = sorted(list(glob.glob(os.path.join(INPUT_DIR_1, 'disparity/TEST/*/*/*/'))))

l = copy.deepcopy(disparity_scenes_train)
for p in l:
    if 'right' in p:
        disparity_scenes_train.remove(p)
        
l = copy.deepcopy(disparity_scenes_test)
for p in l:
    if 'right' in p:
        disparity_scenes_test.remove(p)

disparity_scenes = random.sample(disparity_scenes_train, 1200) + random.sample(disparity_scenes_test, 120)

pool = multiprocessing.Pool(processes=8)
for s in tqdm(disparity_scenes):
    pool.apply_async(proc_one_scene, (s, INPUT_DIR_1, INPUT_DIR_2, OUTPUT_DIR))
    
pool.close()
pool.join()

# for s in tqdm(disparity_scenes):
#     proc_one_scene(s, INPUT_DIR_1, INPUT_DIR_2, OUTPUT_DIR)
```

```python
pi = 3.14159
h = optical_flow.shape[0]
w = optical_flow.shape[1]

flow_degree = np.zeros((h, w))

for i in range(h):
    for j in range(w):
        x_vec = optical_flow[i, j, 0]
        y_vec = optical_flow[i, j, 1]
        d = np.arctan(y_vec / (x_vec+1e-10))
        if x_vec > 0:
            flow_degree[i, j] = d if y_vec > 0 else (2 * pi - d)
        else :
            flow_degree[i, j] = (pi - d) if y_vec > 0 else (pi + d)

flow_degree = (flow_degree - flow_degree.min()) / (flow_degree.max()-flow_degree.min())

flow_img_hsv = np.stack((flow_degree, np.ones((h, w))*0.8, np.ones((h, w))), 2)
flow_img_rgb = hsv2rgb(flow_img_hsv)
plt.imshow(flow_img_rgb)
```
